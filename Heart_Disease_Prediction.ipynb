Import the medical dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection Processing

```python
#loading the csv file pandas dataframe

heart_data=pd.read_csv('/content/heart_disease_data.csv')
```

```python
#print 5 rows of the dataset
heart_data.head()
```

```python
#print last 5 rows of the dataset
heart_data.tail()
```

```python
# the number and colums are present in the dataset
heart_data.shape
```

```python
#getting some input about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
# stastical messaures of the dataset
heart_data.describe()
```

```python
# checking the distribution of target values
heart_data['target'].value_counts()
```

1 represents the Defective heart

0 represents the effective heart


spliting the features

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

spliting the test and train data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

model training

Logestic Regression Model

```python
model = LogisticRegression()
```

```python
# training the logestic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Building predictive system


```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
